```python
##Pandas + Matplotlib Project
"""Accept the class1 percentage class2,class3 as per user
•Create a DataFrame out of this Data out of this 
•Get values from the columns of the DataFrame and plot a Comparison Plot(plot every data in single plot) With different Colors
   You need to compare the data using Line Plot as well as Scatter Plot.
•You need to compare the data using there subplots also"""
```

```python
import pandas as pd
import matplotlib.pyplot as plt
while True :
    try :
        num = int(input(" How many student's marks you want to store :"))
        break
    except :
        print("You have enterd something wrong... Plz enter integer value.")
        
l = []
list1 = []
list2 =[]
list3 = []

for i in range(num):
    name = input(f"Enter the name of student {i+1} ---->")
    l.append(name)
    
    while True :
        try :
            marks1 = float(input(f"Enter the class 1 percentage of {l[i]} ---->"))
            list1.append(marks1)
            break
        except :
            print("Percentage must be in integer or float form... Plz enter again in correct way.")
            
    while True :
        try :
            marks2 = float(input(f"Enter class 2 percentage of {l[i]} ---->"))
            list2.append(marks2)
            break
        except :
            print("Percentage must be in integer or float form... Plz enter again in correct way.")
            
    while True :
        try :
            marks3 = float(input(f"Enter class 3 percentage of {l[i]}"))
            list3.append(marks3)
            break
        except :
            print("Percentage must be in integer or float form... Plz enter again in correct way.")
            
dct = {"name_of_students":l, "class1" : list1 , "class2" : list2 , "class3" : list3}
df = pd.DataFrame(dct)
df


```

```python
#LINE PLOT COMPARISION:
plt.plot(df.name_of_students, df.class1, color ='red')
plt.plot(df.name_of_students, df.class2, color ='blue')
plt.plot(df.name_of_students, df.class3, color ='green')
plt.title("Comparision of marks")
plt.xlabel("---Name of Students---")
plt.ylabel("---Marks---")
plt.legend(["class1","class2","class3"])
plt.show()
```

```python
#SCATTER PLOT COMPARISION:
plt.scatter(df.name_of_students, df.class1, color = 'red')
plt.scatter(df.name_of_students, df.class2, color = 'blue')
plt.scatter(df.name_of_students, df.class3, color = 'green')
plt.title("Comparision of Marks")
plt.xlabel("---Name of Students---")
plt.ylabel("---Marks---")
plt.legend(["class1","class2","class3"])
plt.show()
```

```python
#SUBPLOT COMPARISION:
plt.subplot(2,2,1)
plt.plot(df.name_of_students, df.class1, color = 'red')
plt.xlabel("Students Name")
plt.ylabel("Marks")
plt.legend(['class1'])
plt.show()

plt.subplot(2,2,2)
plt.plot(df.name_of_students, df.class2, color = 'blue')
plt.xlabel("Students Name")
plt.ylabel("Marks")
plt.legend(['class2'])
plt.show()

plt.subplot(2,2,3)
plt.plot(df.name_of_students, df.class3, color = 'green')
plt.xlabel("Students Name")
plt.ylabel("Marks")
plt.legend(['class3'])
plt.show()
```
